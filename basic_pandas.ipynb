#Playing with Dictoinary

```python
import pandas as pd

calories = {"Day 1":1500, "Day 2": 2100, "Day 3": 1000}
mycal = pd.Series(calories)
print(mycal)
print("Previos: ", mycal.loc["Day 3"])
mycal.loc["Day 3"] += 500
print("New: ", mycal.loc["Day 3"])
```

```python
print("Missed diet plan: ", mycal[mycal >= 2000])
print("Sticked to diet plan:", mycal[mycal<2000])
```

DataFrame = 2D Tabular data structure with rowa and columns

```python
data = {
    "Name": ["Mango","Guava","Strawberry","Pineapple"],
    "Price": [50,30,150,70]
}

myData = pd.DataFrame(data, index=["Item 1","Item 2","Item 3","Item 4"])
print(myData)
```

```python
# Add a new column

myData["Color"] = ["Yellow","Green","Red","Orange"]
print(myData)
```

```python
# Add a new row
new_item = pd.DataFrame([{"Name": "Banana","Price": 120,"Color": "Yellow"}],
                        index = ["Item 5"])
myData = pd.concat([myData,new_item])
print(myData)
```

```python
# Adding new items
new_items = pd.DataFrame([{"Name": "Lichi","Price": 100,"Color": "Red"},
                          {"Name": "Watermelon","Price": 150,"Color": "Green"},
                          {"Name": "Apple","Price": 250,"Color": "Red"}],
                        index = ["Item 6","Item 7","Item 8"])
myData = pd.concat([myData,new_items])
print(myData)
```

```python
# Filtering
tall_pokemon = c_data[c_data["Height"] >= 3]
heavy_pokemon = c_data[c_data["Weight"] > 100]
legendary_pokemon = c_data[c_data["Legendary"] == True]
water_pokemon = c_data[(c_data["Type1"] == "Water") | (c_data["Type2"] == "Water")]
print("Tall pokemon:",tall_pokemon)
print("Heavy pokemon:",heavy_pokemon)
print("Legendary pokemon:",legendary_pokemon)
print("Water pokemon:",water_pokemon)
```

```python
# Aggregate Function = used to summarize set of values

# Whole dataframe
print("Mean:",c_data.mean(numeric_only=True))
print("Sum:",c_data.sum(numeric_only=True))
print("Min:",c_data.min(numeric_only=True))
print("Max:",c_data.max(numeric_only=True))
print("Count:",c_data.count(numeric_only=True))

# Single Column
print("Mean of height:",c_data["Height"].mean())
print("Sum of height:",c_data["Height"].sum())

```

```python
# Group by
group = c_data.groupby("Type1")
print(group["Height"].mean().head())
```

```python
# Data Cleaning = Fixing/Removing incorrect or incomplete info
# Drop irrelevant column
c_data = c_data.drop(columns=["Legendary"])
print(c_data.head())
```

```python
# c_data = c_data.dropna(subset=["Type2"])
c_data = c_data.fillna({"Type2": "None"})
print(c_data.head())
```

```python
# Fix inconsistance values
c_data["Type1"] = c_data["Type1"].replace({"Grass":"GRASS","Fire":"FIRE"})
print(c_data.head())
```

```python
# Standardize text
c_data["Type2"] = c_data["Type2"].str.upper()
print(c_data.head())
```
